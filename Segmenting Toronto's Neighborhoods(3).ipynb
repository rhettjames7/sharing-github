# Section 1, Question 1: Segmenting Toronto's Neighborhood

### First step is to scrape Toronto postcodes from a Wikipedia page here: https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M

```python
import requests
import pandas as pd
website_url = requests.get('https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M').text
from bs4 import BeautifulSoup
soup = BeautifulSoup(website_url,'lxml')
```

```python
# Pull specifically the M table from the Wikipedia page
theTable = soup.find('table', class_='wikitable sortable')
```

```python
# We want the 3 columns of this table converted to a Pandas dataframe
# Since the <td> tags are consecutive in each row, we will use BeautifulSoup to extract
# these values from the wiki table by column, where we start at the first <td> and gain
# the next value every 3rd time. Since BeautifulSoup's .text function returns a string,
# we must convert it back to a list value each for iteration via .split()

column1 = []
for column1raw in theTable.findAll('td')[0::3]:
    #print(column1raw.text)
    column1 += column1raw.text.split()
#print(column1)
```

```python
column2 = []
for column2raw in theTable.findAll('td')[1::3]:
    #print(column2raw.text)
    column2.append(column2raw.text) #.append here to cater to the data
#print(column2)
```

```python
column3 = []
for column3raw in theTable.findAll('td')[2::3]:
    #print(column3raw.text)
    column3.append(column3raw.text[:-1]) #-1 eliminates the \n at the end of each value from this column
#print(column3)
```

## Question 2: Scrape the Wikipedia page
### Question 3, bullet 1: Use 3 columns

```python
fullTable = {'Postcode': column1, #Convert lists into columns with names via dict
             'Borough': column2,
             'Neighbourhood': column3}
df = pd.DataFrame(fullTable)
df
```

### Question 3, Bullet 2: Ignore cells with a borough that is Not assigned

```python
# Next we drop all boroughs that are Not assigned
df = df[df.Borough != 'Not assigned']
df.head()
```

### Question 3, Bullet 3: Combine neighbourhoods via ',' that have a matching Postcode and Borough

```python
result = df.groupby(['Postcode', 'Borough'], as_index=True)['Neighbourhood'].apply(','.join)
result.head()
#convert result back to Pandas DataFrame
df2 = pd.DataFrame(result)
df2 = df2.reset_index()
df2
```
